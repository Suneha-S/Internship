```python
#counting number of duplicate values
df = pd.read_csv(r'C:\Users\l\OneDrive\Desktop\Medical Appointment No Shows.csv')
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
#droping duplicates
df_cleaned = df.drop_duplicates()
df_cleaned.to_csv(r'C:\Users\l\OneDrive\Desktop\Medical Appointment No Shows.csv', index=False)
print(df_cleaned.head())
```

```python
#Standardizing values and printing it 
df = pd.read_csv(r'C:\Users\l\OneDrive\Desktop\Medical Appointment No Shows.csv')
df['Gender'] = df['Gender'].str.lower().str.strip()
df['Gender'] = df['Gender'].replace({'f': 'female', 'm': 'male'})
df['Neighbourhood'] = df['Neighbourhood'].str.strip().str.title()
df['No-show'] = df['No-show'].str.strip().str.capitalize()
df.head()
```

```python
#Converting date columns to a consistent datetime format
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'], errors='coerce')
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'], errors='coerce')
df.head()
```

```python
# Cleaning column names lowercase, replace spaces and hyphens with underscores
df = pd.read_csv(r'C:\Users\l\OneDrive\Desktop\Medical Appointment No Shows.csv')
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('-', '_')
df.columns
```

```python
#converting date format
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['ScheduledDay'] 
```

```python
#converting date format
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
df['AppointmentDay'] 
```

```python
#saving the processed file 
output_path = r"C:\Users\l\OneDrive\Desktop\processed_Medical Appointment No Shows.csv"
df.to_csv(output_path, index=False)
print("Cleaned file saved to Desktop!")
```
